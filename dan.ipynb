Scatter Plot

```python
plt.figure(figsize=(10,2))
plt.scatter(df['Roll'],df['Age'])
plt.title("class vs student")
plt.xlabel("Age")
plt.ylabel("Roll")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.plot(df['Roll'],df['Age'],marker=0)
plt.title("class vs student")
plt.xlabel("Age")
plt.ylabel("Roll")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.bar(df['Roll'],df['Age'])
plt.title("class vs student")
plt.xlabel("Age")
plt.ylabel("Roll")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.hist(df['Age'],bins=6)
plt.title("class vs stude,nt")
plt.xlabel("Age")
plt.ylabel("Roll")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.boxplot(df['Age'])
plt.title("class vs student")
plt.grid(True)
plt.show()
```
